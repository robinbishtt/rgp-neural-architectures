## FSS Collapse

```python
from src.scaling.fss_analysis import DataCollapser, CriticalExponentFitter
import numpy as np

# Synthetic FSS data
widths = np.array([64, 128, 256, 512], dtype=float)
xi_values = np.array([5.0, 15.0, 50.0, 100.0, 200.0])
nu_true, xi_c_true = 1.0, 20.0

rng = np.random.default_rng(42)
acc_matrix = np.array([
    [0.5 + 0.4/(1+np.exp(-((xi-xi_c_true)*N**(1/nu_true))/5)) + rng.normal(0,0.01)
     for N in widths]
    for xi in xi_values
])

fitter = CriticalExponentFitter()
nu_best, ci_lo, ci_hi = fitter.fit(xi_values, widths, acc_matrix, n_bootstrap=100)
print(f"Fitted nu = {nu_best:.3f} (true={nu_true})")
print(f"95% CI: [{ci_lo:.3f}, {ci_hi:.3f}]")
print(f"Contains nu=1.0 (mean-field): {ci_lo <= 1.0 <= ci_hi}")
```

Verifies c^(ℓ)_ij = c^(0)_ij · χ^ℓ. The depth scale k_c = -1/log(χ).

## Mean-Field Correlation Propagation

```python
from src.proofs.theorem2_exponential_decay import propagate_correlation, run_all_verifications

results = run_all_verifications()
print("Theorem 2 results:", results)

# Show decay for different chi values
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax, chi in zip(axes, [0.5, 0.8, 0.99]):
    c = propagate_correlation(0.9, chi, 50)
    layers = np.arange(len(c))
    k_c = -1.0 / np.log(chi)
    ax.plot(layers, c, "o", ms=3, label="Simulated")
    ax.plot(layers, 0.9 * np.exp(-layers/k_c), "-", label=f"exp(-k/{k_c:.1f})")
    ax.set_title(f"χ={chi}, k_c={k_c:.1f}"); ax.legend()
    ax.set_xlabel("Layer k"); ax.set_ylabel("c¹²(k)")
plt.tight_layout(); plt.show()
```

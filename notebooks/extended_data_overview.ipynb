## Generate All Extended Data Figures

```python
import subprocess, sys
from pathlib import Path

ed_scripts = [
    "figures/extended_data/run_extended_figure1.py",
    "figures/extended_data/run_extended_figure2.py",
    "figures/extended_data/run_extended_figure3.py",
]

output_dir = Path("/tmp/extended_data_nb/")
output_dir.mkdir(parents=True, exist_ok=True)

for script in ed_scripts:
    script_path = Path("..") / script
    if script_path.exists():
        result = subprocess.run(
            [sys.executable, str(script_path), "--fast-track",
             "--output", str(output_dir)],
            capture_output=True, text=True, cwd=".."
        )
        print(f"{script}: {'OK' if result.returncode == 0 else 'FAIL'}")
        if result.stdout: print(" ", result.stdout.strip()[:80])
    else:
        print(f"  {script}: not found (run from repo root)")
```

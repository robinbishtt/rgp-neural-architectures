# Phase Diagram: Ordered / Critical / Chaotic Phases

Reproduces the (σ_w, σ_b, χ) phase diagram. Critical surface: χ₁ = 1.



```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
import torch

from src.utils.seed_registry import SeedRegistry
from src.utils.device_manager import DeviceManager

SeedRegistry.get_instance().set_master_seed(42)
device = DeviceManager.get_instance().get_device()
print(f'Device: {device}')
```

## χ₁ as a function of (σ_w, σ_b)

```python
from src.core.correlation.two_point import chi1_gauss_hermite

sigma_w_vals = np.linspace(0.5, 3.0, 60)
sigma_b_vals = np.linspace(0.0, 1.0, 40)
SW, SB = np.meshgrid(sigma_w_vals, sigma_b_vals)
CHI = np.zeros_like(SW)

for i in range(SW.shape[0]):
    for j in range(SW.shape[1]):
        CHI[i, j] = chi1_gauss_hermite(SW[i, j]**2, "tanh")

fig, ax = plt.subplots(figsize=(7, 5))
cf = ax.contourf(SW, SB, CHI, levels=50, cmap="RdBu_r")
cs = ax.contour(SW, SB, CHI, levels=[1.0], colors="black", linewidths=2)
ax.clabel(cs, fmt="χ₁=1", fontsize=10)
plt.colorbar(cf, ax=ax, label="χ₁")
ax.plot(1.4, 0.3, "r*", ms=15, label="Paper critical: (1.4, 0.3)")
ax.set_xlabel("σ_w"); ax.set_ylabel("σ_b")
ax.set_title("Phase Diagram: χ₁(σ_w, σ_b) for tanh activation")
ax.legend(); plt.tight_layout(); plt.show()

# Paper values
chi_paper = chi1_gauss_hermite(1.4**2, "tanh")
print(f"chi1 at (sigma_w=1.4, sigma_b=0.3): {chi_paper:.4f}")
print(f"Note: chi1 < 1 means near-critical ordered phase (paper intended chi1 ≈ 1)")
```

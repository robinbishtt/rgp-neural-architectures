```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
k = np.arange(0, 20)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].plot(k, 5.0 * np.exp(-k / 8.0), color='#2166ac', linewidth=2)
axes[0].set_title('Ordered Phase (chi1 < 1)')
axes[0].set_xlabel('Layer index k')
axes[0].set_ylabel('xi(k)')

axes[1].plot(k, 5.0 * np.exp(-k / 1.5), color='#d6604d', linewidth=2)
axes[1].set_title('Chaotic Phase (chi1 > 1)')
axes[1].set_xlabel('Layer index k')
axes[1].set_ylabel('xi(k)')

axes[2].plot(k, 5.0 / (1.0 + k * 0.4), color='#1a9641', linewidth=2)
axes[2].set_title('Critical Point')
axes[2].set_xlabel('Layer index k')
axes[2].set_ylabel('xi(k)')

plt.tight_layout()
plt.savefig('nb01_phase_overview.pdf', bbox_inches='tight', dpi=150)
plt.show()
```

```python
def fisher_correlation_length(W):
    n  = W.shape[0]
    F  = (W.T @ W) / n
    ev = np.clip(np.linalg.eigvalsh(F), 1e-10, None)
    return float(1.0 / np.sqrt(np.mean(1.0 / ev)))

rng = np.random.default_rng(42)
n, m = 256, 20

for label, sigma_w in [
    ('Ordered  (sigma_w = 0.80)', 0.8),
    ('Critical (sigma_w ~ 1.48)', 1.0 / np.sqrt(0.456)),
    ('Chaotic  (sigma_w = 2.50)', 2.5),
]:
    xi_profile = [
        fisher_correlation_length(
            rng.standard_normal((n, n)) * sigma_w / np.sqrt(n)
        )
        for _ in range(m)
    ]
    print(f'{label}:  xi(0)={xi_profile[0]:.3f},  xi(19)={xi_profile[-1]:.4f}')
```

```python
xi_data   = np.linspace(2.0, 100.0, 300)
xi_target = 1.0

fig, ax = plt.subplots(figsize=(7, 4))
for k_c in [1.0, 2.0, 5.0]:
    ax.plot(xi_data, k_c * np.log(xi_data / xi_target),
            label=f'k_c = {k_c:.0f}', linewidth=2)

ax.set_xlabel('xi_data')
ax.set_ylabel('L_min')
ax.set_title('H2: Logarithmic Depth Scaling')
ax.legend()
plt.tight_layout()
plt.savefig('nb01_depth_prescription.pdf', bbox_inches='tight', dpi=150)
plt.show()
```

## Critical Points for Different Activations

```python
from src.core.correlation.two_point import chi1_gauss_hermite, critical_sigma_w2

activations = ["tanh", "relu", "gelu"]
print(f"{'Activation':<10} {'sigma_w*':>10} {'chi1 at crit':>15} {'xi_depth':>12}")
print("-" * 50)
for act in activations:
    sw2_star = critical_sigma_w2(act)
    chi1_c = chi1_gauss_hermite(sw2_star, act)
    xi_d = -1/np.log(max(chi1_c, 1e-10)) if chi1_c < 1 else float("inf")
    print(f"{act:<10} {sw2_star**0.5:>10.4f} {chi1_c:>15.6f} {xi_d:>12.1f}")
```
